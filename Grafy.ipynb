```python
import json
import matplotlib.pyplot as plt
from IPython.display import display
%pylab inline
#plt.xkcd()
def sel(x, index):
    return map(lambda xx: xx[index], x)
min_count_filter = 1000

if False:
    data = json.load(open('stat.trf_cover.stat'))
    indels = map(lambda x: (x[0], x[1], sum(x[2])),data['indels'])
    ident = map(lambda x: (x[0], x[1], sum(x[2])),data['identity'])

    indels = filter(lambda x: sum(x[2]) > min_count_filter, indels)
    ident = filter(lambda x: sum(x[2]) > min_count_filter, ident)

    plt.plot(sel(ident, 0), sel(ident, 1), label="Sequence Identity (human-chicken)")
    plt.plot(sel(indels, 0), sel(indels, 1), label="Indel ratio (human-chicken)")

data = json.load(open('stat.trf_cover.stat.orig'))
indels = map(lambda x: (x[0], x[1], sum(x[2])),data['indels'])
ident = map(lambda x: (x[0], x[1], sum(x[2])),data['identity'])

indels = filter(lambda x: sum(x[2]) > min_count_filter, indels)
ident = filter(lambda x: sum(x[2]) > min_count_filter, ident)

plt.plot(sel(ident, 0), sel(ident, 1), label="Sequence Identity (human-dog)")
plt.plot(sel(indels, 0), sel(indels, 1), label="Indel ratio (human-dog)")

data = json.load(open('stat.trf_cover.stat.13'))
indels = map(lambda x: (x[0], x[1], sum(x[2])),data['indels'])
ident = map(lambda x: (x[0], x[1], sum(x[2])),data['identity'])

indels = filter(lambda x: sum(x[2]) > min_count_filter, indels)
ident = filter(lambda x: sum(x[2]) > min_count_filter, ident)

plt.plot(sel(ident, 0), sel(ident, 1), label="Sequence Identity (human-dog realigned)")
plt.plot(sel(indels, 0), sel(indels, 1), label="Indel ratio (human-dog realigned)")

#plt.plot([0,0.0],[0,1.0],color='gray')
plt.xlabel('distance from repeat border')
#plt.ylabel('ratio')
plt.legend(bbox_to_anchor=(2.1, 1.05))
plt.xlim([-120, 120])
#plt.ylim([0.6, 0.8])
plt.ylim([0.1, 0.3])

plt.show()

```
