```python
import json
import matplotlib.pyplot as plt
from IPython.display import display
%pylab inline
#plt.xkcd()

def parse_lines(x):
    d = dict()
    for line in x:
        line = line.strip().split(' ')
        d[line[0]] = map(lambda x: 100.0-float(x), line[1:])
    return d

dashes=[[], [8, 4], [8, 4, 2, 4], [2, 2]]
def draw_plot(data, selections=None, captions=None, filename=None):
    if selections == None:
        selections = data.keys()
    if captions == None:
        captions = selections
    for i in range(len(selections)):
        key = selections[i]
        label = captions[i]
        dat = data[key]
        line, = plt.plot(range(1, len(dat) + 1), dat, 'k-' ,label=label)
        line.set_dashes(dashes[i]) 
    plt.xlabel('Distance from repeat border')
    plt.ylabel('Error rate')
    plt.ylim(2, 18)
    plt.xlim(1, 20)
    plt.xticks(range(2, 21, 2))
    plt.legend()#bbox_to_anchor=(2.1, 1.05))
    if filename != None:
        pylab.savefig(filename)
    plt.show()

    

with open('distance_data.cvs') as f:
    data = parse_lines(list(f))
#data
draw_plot(data, 
          [
           'repeatRealignerBrona', 
           'sff_block_trfhmm2_cor_postprocessed',
           'simpleModelViterbi', 
           'simpleModelReplaceNonOv_cor',
          ], [
           'SFF marginalized', 
           'SFF block', 
           '3-state Viterbi',
           '3-state Model, best repeat masking'
          ],
          'figures/error_graph_overview.pdf')
draw_plot(data, 
          [
           'sff_block_viterbi_trfhmm2_cor_postprocessed', 
           'sff_block_trfhmm2_cor_postprocessed',
           'sff_block_viterbi_original_cor',
           'sff_block_original_cor', 
          ], [
           'SFF block Viterbi', 
           'SFF block',  
           'SFF block Viterbi original',
           'SFF block original',
          ],
          'figures/error_graph_sffblock.pdf')

draw_plot(data, 
          [
           'repeatRealignerBrona', 
           'repeatRealignerBronaOriginal',
           'repeatRealignerBronaChicken', 
           'repeatRealignerBronaPerturbed',
          ], [
           'SFF marginalized', 
           'SFF marginalized original', 
           'SFF marginalized chicken',
           'SFF marginalized perturbed',
          ],
          'figures/error_graph_marginalized.pdf')
draw_plot(data, 
          [
           'sff_block_trfhmm2_cor_postprocessed', 
           'sff_block_original_cor_postprocessed',
           'sff_block_high2_cor_postprocessed', 
           'sff_block_high2_orig_cor',
          ], [
           'SFF block', 
           'SFF block original', 
           'TANTAN block',
           'TANTAN block origina',
          ],
          'figures/error_graph_sffvstantan.pdf')
draw_plot(data, 
          [
           'simpleModelReplaceNonOv_cor', 
           'simpleModelReplace_cor',
           'simpleModelReplaceHMM_cor', 
           'simpleModelReplaceOrig_cor',
          ], [
           '3-state posterior, TRF non-overlapping masking', 
           '3-state posterior, TRF masking ', 
           '3-state posterior, Sunflower masking',
           '3-state posterior, Correct masking',
          ],
          'figures/error_graph_3statemasking.pdf')
draw_plot(data, 
          [
           'muscle',
           'context', 
           'simpleModelViterbi', 
           'simpleModelReplaceNonOv_cor',
          ], [
           'muscle', 
           'context',
           '3-state Viterbi', 
           '3-state posterior, TRF non-overlapping masking',
          ],
          'figures/error_graph_other.pdf')
```
